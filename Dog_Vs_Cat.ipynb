```python
# creating labels for resized images of dogs  and cats
file_names=os.listdir('/content/image resized/')
# cat=0 && dog=1
labels=[]

for i in range(2000):
  filename=file_names[i]
  label=filename[0:3]
  if(label=='dog'):
    labels.append(1)
  else:
    labels.append(0)

```

```python
x=dog_cat_images
y=np.asarray(labels)
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2, random_state=2)
```

```python
print(x.shape,x_train.shape, x_test.shape)
```

```python
x_train=x_train/255
x_test=x_test/255
```

```python
print(x_train, x_test)
```

```python
import tensorflow as tf
import tensorflow_hub as hub
```

```python
# mobilenet='https://tfhub.dev/google/tf2-preview/mobilenet_v2/feature_vector/4'
mobilenet='https://www.kaggle.com/models/google/mobilenet-v2/tensorFlow2/tf2-preview-feature-vector/4?tfhub-redirect=true'
```

```python
pretrained_model=hub.KerasLayer(mobilenet, input_shape=(224,224,3),trainable=False)
```

```python
# num_classes=2

# model=tf.keras.Sequential([
#     pretrained_model,
#     tf.keras.layers.Dense(num_classes)
# ])

# model.summary()
import tensorflow as tf
import tensorflow_hub as hub

# Use the correct TensorFlow Hub URL for MobileNetV2
mobilenet_url = 'https://tfhub.dev/google/tf2-preview/mobilenet_v2/feature_vector/4'

# Load the pre-trained MobileNetV2 model from TensorFlow Hub
pretrained_model = hub.KerasLayer(mobilenet_url, input_shape=(224, 224, 3), trainable=False)

num_classes = 2

# Build the model using the Functional API
inputs = tf.keras.Input(shape=(224, 224, 3))
X = pretrained_model(inputs)
outputs = tf.keras.layers.Dense(num_classes, activation='softmax')(X)
model = tf.keras.Model(inputs=inputs, outputs=outputs)

model.summary()
```

```python
model.compile(optimizer='adam', loss=tf.keras.losses.SparseCategoricalCrossentropy(from_logits=True),
              metrics=['acc'])
```

```python
model.fix(x_train,y_train, epochs=5)
```

```python
score,acc=model.evaluate(x_test,y_test)
```
